```python

def isSafe(mat, r1, c1):
    for i1 in range(r1):
        if mat[i1][c1] == 'N':
            return False

    (i1, j1) = (r1, c1)
    while i1 >= 0 and j1 >= 0:
        if mat[i1][j1] == 'N':
            return False
        i1 = i1 - 1
        j1 = j1 - 1

    (i1, j1) = (r1, c1)
    while i1 >= 0 and j1 < len(mat):
        if mat[i1][j1] == 'N':
            return False
        i1 = i1 - 1
        j1 = j1 + 1
 
    return True
 
 
def sol(mat):
    for r1 in mat:
        print(str(r1).replace(',', '').replace('\'', ''))
    print()
 
def mQueen(mat, r1):
    if r1 == len(mat):
        sol(mat)
        return
    for i1 in range(len(mat)):

        if isSafe(mat, r1, i1):
            mat[r1][i1] = 'N'
            mQueen(mat, r1 + 1)

            mat[r1][i1] = '*'
 
 

N1 = 4
mat = [['*' for x in range(N1)] for y in range(N1)]

start = time.time()
mQueen(mat, 0)
end = time.time()
backtracking.append(end - start)
```

```python
import math
res1 = []
def boar(board1, row1, rowmask1,ldmask1, rdmask1):
	n = len(board1)
	all_rows = (1 << n) - 1
	if (rowmask1 == all_rows):
		v1 = []
		for i1 in board1:
			for j1 in range(len(i1)):
				if i1[j1] == 'N':
					v1.append(j1+1)
		res1.append(v1)
	s = all_rows & (~(rowmask1 | ldmask1 | rdmask1))

	while (s > 0):
		p1 = s & (-s)
		col = (int)(math.log(p1)/math.log(2))
		board1[row1][col] = 'N'
		boar(board1, row1+1, rowmask1 | p1,
				(ldmask1 | p1) << 1, (rdmask1 | p1) >> 1)
		s = s & (s-1)
		board1[row1][col] = ' '
def sol(board1):
	for row1 in board1:
		print("|" + "|".join(row1) + "|")

```

```python
def m():

	n = 4
	board1 = []

	for i1 in range(n):
		row1 = []
		for j1 in range(n):
			row1.append(' ')
		board1.append(row1)

	rowmask1 = 0
	ldmask1 = 0
	rdmask1 = 0
	row1 = 0

	res1.clear()
	boar(board1, row1, rowmask1, ldmask1, rdmask1)
	res1.sort()
	print(res1)

start = time.time()
m()
end = time.time()
Modified_backtring.append(end - start)
```

```python

import random

def nqueens(nr):
	show(min_conflicts(list(range(nr)), nr), nr)

def show(soln, nr):
	for i in range(nr):
		row = ['~'] * nr
		for col in range(nr):
			if soln[col] == nr - 1 - i:
				row[col] = 'N'
		print(''.join(row))

def min_conflicts(soln, nr, iters=1000):
	def random_pos(li, filt):
		return random.choice([i for i in range(nr) if filt(li[i])])

	for k in range(iters):
		confs = find_conflicts(soln, nr)
		if sum(confs) == 0:
			return soln
		col = random_pos(confs, lambda elt: elt > 0)
		vconfs = [hits(soln, nr, col, row) for row in range(nr)]
		soln[col] = random_pos(vconfs, lambda elt: elt == min(vconfs))
	raise Exception("Incomplete solution: try more iterations.")

def find_conflicts(soln, nr):
	return [hits(soln, nr, col, soln[col]) for col in range(nr)]

def hits(soln, nr, col, row):
	total = 0
	for i in range(nr):
		if i == col:
			continue
		if soln[i] == row or abs(i - col) == abs(soln[i] - row):
			total += 1
	return total

nqueens(4)
```
